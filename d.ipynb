```python
import numpy as np
import pandas as pd
from chembl_webresource_client.new_client import new_client

drug = new_client.drug
drugs = drug.filter(warning_type='Withdrawn').only("molecule_chembl_id")
df = pd.DataFrame(drugs)

mechanism_of_action = []

i = 0
for el in df['molecule_chembl_id']:
    result = new_client.mechanism.filter(molecule_chembl_id = el).only("mechanism_of_action")
    
    if result[0] !=None:
        mechanism_of_action.append(result[0]['mechanism_of_action'])
    i = i + 1

    if i%200 == 0:
        print(f'Molecule counting : {i}')
```

```python
from collections import Counter

count=Counter(mechanism_of_action)
print("Order of the most equal action mechanism :" ,count)
```

```python
count=Counter(mechanism_of_action)
print("The most equal action mechanism : " , count.most_common(1)) 
```

```python
count=Counter(mechanism_of_action)
count=Counter(mechanism_of_action)
print("Three of the most equal action mechanisms :" , count.most_common(3))
```

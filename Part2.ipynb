```python
from chembl_webresource_client.new_client import new_client
import numpy as np
import pandas as pd

drug = new_client.drug
d0 = drug.filter(withdrawn_reason__isnull = False).only(['molecule_structures', 'synonyms','molecule_chembl_id', 'molecule_properties','num_lipinski_ro5_violations', 'num_ro5_violations'])
```

```python
d0[0]
```

```python
d0[0]['molecule_properties'].keys()
```

```python
#we can check 'num_ro5_violations', 'hba' and 'hbd' in molecule_properties col.
```

```python
df_d0 = pd.DataFrame(d0)
df_d0
```

```python
del df_d0['applicants']
del df_d0['atc_code_description']
del df_d0['research_codes']
df_d0
```

```python
d1 = drug.filter(withdrawn_reason__isnull = False).only('molecule_structures')

canonical_smiles=[]

for i in range(144):
  try:
    canonical_smiles.append(d1[i]['molecule_structures']['canonical_smiles'])
  except TypeError:
    canonical_smiles.append(None)
```

```python
df_d0['molecule_structures'] = canonical_smiles
```

```python
df_d0.head()
```

```python
df_d0['molecule_properties'][4]['num_ro5_violations'] #num_ro5_violations, hba, hbd
```

```python
hba=[]

for i in range(144):
  try:
    hba.append(df_d0['molecule_properties'][i]['hba'])
  except TypeError:
    hba.append(None)
```

```python
hbd=[]

for i in range(144):
  try:
    hbd.append(df_d0['molecule_properties'][i]['hbd'])
  except TypeError:
    hbd.append(None)
```

```python
num_ro5_violations=[]

for i in range(144):
  try:
    num_ro5_violations.append(df_d0['molecule_properties'][i]['num_ro5_violations'])
  except TypeError:
    num_ro5_violations.append(None)
```

```python
del df_d0['molecule_properties']
```

```python
df_d0
```

```python
df_d0.insert(loc = 1, column='hba',value=hba)
df_d0.insert(loc = 1, column='hbd',value=hbd)
df_d0.insert(loc = 1, column='num_ro5_violations',value=num_ro5_violations)
```

```python
df_d0
```

```python
df_d0 = df_d0.rename(columns={'molecule_structures': 'canonical_smiles', 'synonyms': 'FDA name'})
```
